# Visualizing Gender Gap

## Objective

- The data is clean and needs no  further cleaning (LOL they lied when they said data science is 80% data cleaning :) )
- We will be doing two plots but with repitations that will introduce some more improvements>

## Task outline
1. Import the required libraries we need
2. Load the data set into pandas (Yay! Pandas for life )
3. display the dataset (no need to look for few rows, as this is small dataset)
4. Display Gender Gap for Stem subjects
    - Identify the Stem subjects and put them in a list
    - Plot the percentage of women  each stem subject should have its own plot in an 2 row with 3 column axes
    - Now add the men percentage into the same plots above 
    - lets add the labels to the plots to show what is women and what is men

    - Remove the legend on the second row, because they are in first row the second row does not need them
    - Remove the ticks on the axes!
5. Display Gender Gap for all degree categories

    - Add title to each plot

```python
#import the libraries needed

import pandas as pd
import matplotlib.pyplot as plt
from datetime import *
import numpy as np
import time 
```

```python
#load data into pandas
degrees_women = pd.read_csv('percent-bachelors-degrees-women-usa.csv') 
```

```python
#Displa dataset
degrees_women
```

```python
#from the data above lets create a list of stem subjects
stem_subjects = ['Biology','Computer Science','Engineering','Math and Statistics','Physical Sciences','Psychology']
len(stem_subjects)
```

```python
#for each subject plot the percentage of women
fig = plt.figure(figsize=(20,5))
for i in range(len(stem_subjects)):
    ax = fig.add_subplot(2, 3, i+1)
    ax.plot(degrees_women['Year'],degrees_women[stem_subjects[i]])
    
```

```python
#for each subject plot the percentage of women and men in the same plot

fig =plt.figure(figsize=(20,10))

for i in range(len(stem_subjects)):
    ax = fig.add_subplot(2,3, i+1)
    ax.plot(degrees_women['Year'], degrees_women[stem_subjects[i]])
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_subjects[i]]))
```

```python
#lets add the title of each plot 
fig =plt.figure(figsize=(20,10))

for i in range(len(stem_subjects)):
    ax = fig.add_subplot(2,3, i+1)
    ax.plot(degrees_women['Year'], degrees_women[stem_subjects[i]])
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_subjects[i]]))
    ax.set_title(stem_subjects[i])
```

```python
#lets add the names to the plots to identify them
fig = plt.figure(figsize=(20,10))

for i in range(len(stem_subjects)):
    ax = fig.add_subplot(2,3,i+1)
    ax.plot(degrees_women['Year'], degrees_women[stem_subjects[i]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_subjects[i]]), label = 'Men')
    ax.set_title(stem_subjects[i])
    ax.legend()
    
```

```python
#lets remove the axes; this will enable the important data to be seen better
fig = plt.figure(figsize=(20,10))

for i in range (len(stem_subjects)):
    ax = fig.add_subplot(2,3,i+1)
    
    ax.plot(degrees_women['Year'], degrees_women[stem_subjects[i]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_subjects[i]]), label = 'Men')
    ax.set_title(stem_subjects[i])
    ax.legend()
    
    for keys,spine in ax.spines.items():
        spine.set_visible(False)
  
```

```python
#lets remove the  the ticks
fig = plt.figure(figsize=(20,10))

for i in range (len(stem_subjects)):
    ax = fig.add_subplot(2,3,i+1)
    
    ax.plot(degrees_women['Year'], degrees_women[stem_subjects[i]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_subjects[i]]), label = 'Men')
    
    for keys,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_title(stem_subjects[i])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    if i == 0 or i == 1 or i == 2:
        ax.legend()
    

plt.show()    
```

```python
#lets group the degrees into the categories that we want.
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
#lets now plot the gender gaps

fig = plt.figure(figsize=(20,10))

#lets plot the first column
for i in range(0,18,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int(i/3)
    
    ax.plot(degrees_women['Year'], degrees_women[stem_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_cats[cat_index]]), label = 'Men')

#lets plot the second column 
for i in range (1,16,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int((i-1)/3)
    
    ax.plot(degrees_women['Year'], degrees_women[lib_arts_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[lib_arts_cats[cat_index]]), label = 'Men')
    
#lets plot the Third column
for i in range (2,20,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int((i-2)/3)
    
    ax.plot(degrees_women['Year'], degrees_women[other_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[other_cats[cat_index]]), label = 'Men')
```

```python
#let add the plot title of each plot

fig = plt.figure(figsize=(20,10))

#lets plot the first column
for i in range(0,18,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int(i/3)
    
    ax.plot(degrees_women['Year'], degrees_women[stem_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_cats[cat_index]]), label = 'Men')
    ax.set_title(stem_cats[cat_index]) #title of each plot on the first column added here 

#lets plot the second column 
for i in range (1,16,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int((i-1)/3)
    
    ax.plot(degrees_women['Year'], degrees_women[lib_arts_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[lib_arts_cats[cat_index]]), label = 'Men')
    ax.set_title(lib_arts_cats[cat_index]) #title of each plot on the second column added here
    
#lets plot the Third column
for i in range (2,20,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int((i-2)/3)
    
    ax.plot(degrees_women['Year'], degrees_women[other_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[other_cats[cat_index]]), label = 'Men')
    ax.set_title(other_cats[cat_index])#title of each plot on the third column added here
```

```python
#lets remove the axes to make the data display clean

fig = plt.figure(figsize=(20,10))

#lets plot the first column
for i in range(0,18,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int(i/3)
    
    ax.plot(degrees_women['Year'], degrees_women[stem_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[stem_cats[cat_index]]), label = 'Men')
    ax.set_title(stem_cats[cat_index]) #title of each plot on the first column added here 
    
    for keys,spine in ax.spines.items(): #removing the axes
        spine.set_visible(False)

#lets plot the second column 
for i in range (1,16,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int((i-1)/3)
    
    ax.plot(degrees_women['Year'], degrees_women[lib_arts_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[lib_arts_cats[cat_index]]), label = 'Men')
    ax.set_title(lib_arts_cats[cat_index]) #title of each plot on the second column added here
    
    for keys,spine in ax.spines.items(): #removing the axes
        spine.set_visible(False)
    
#lets plot the Third column
for i in range (2,20,3):
    ax = fig.add_subplot(6,3,i+1)
    cat_index = int((i-2)/3)
    
    ax.plot(degrees_women['Year'], degrees_women[other_cats[cat_index]], label = 'Women')
    ax.plot(degrees_women['Year'], 100-(degrees_women[other_cats[cat_index]]), label = 'Men')
    ax.set_title(other_cats[cat_index])#title of each plot on the third column added here
    
    for keys,spine in ax.spines.items():#removing the axes
        spine.set_visible(False)
```
